**Metodo de minimos cuadrados**

```python
# Librerias
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

Vamos a desarrollar un código para resolver el ejercicio por el método de mínimos cuadrados.  

```python
x = sp.symbols('x') # Definimos las variables que vamos a usar
a, b = sp.symbols('a b')
u_n = 2*x + a*x**2 + b*x**3 # Son las condiciones del problema

sp.pprint(u_n) # imprimimos la funcion para ver que todo bien

u_n_2 = sp.diff(u_n, x, 2)  # Segunda derivada

E = u_n_2 + u_n - x # Es el error
```

```python
# Aplicamos minimos cuadrados
J = sp.integrate(E**2, (x, 0, 1))

eq1 = sp.diff(J, a) # Minimiza las derivadas
eq2 = sp.diff(J, b)

sol = sp.solve([eq1, eq2], (a, b)) # Aqui se resuelven las ecuaciones para encontrar los coeficientes

print("\nSoluciones")
print(sol)

u_aproximada = u_n.subs(sol) # En esta parte obtenemos la solucioon final

print("Solucion aprox")
sp.pprint(u_aproximada)

u_exacta = sp.sin(x) + x # Vamos a comparar las soluciones, empezando con la exacta
```

```python
# Errores

u_exacta = sp.sin(x) + x

error_cuadratico = (u_exacta - u_aproximada)**2 # calcula el error global
error_L2 = sp.sqrt(sp.integrate(error_cuadratico, (x, 0, 1)))

val_exacto = u_exacta.subs(x, 1).evalf() # calcula el error puntual
val_aprox = u_aproximada.subs(x, 1).evalf()
error_puntual = abs(val_exacto - val_aprox)

print(f"Error Global : {error_L2.evalf():.8f}") # Esto es  para comparar amas soluciones y ver que tan bueno es el metodo xxd
print(f"Error en x=1 : {error_puntual:.8f}")
```

Notemos que el error es muy pequeño, yo diria que es nulo porque practicamente es.
