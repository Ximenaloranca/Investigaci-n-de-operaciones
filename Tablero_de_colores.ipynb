Tenemos un tablero con 4 sectores; se lanzan dos monedas y el numero de caras que salgan nos dice cuantos sectores se avanzan en el sentido contrario. Nosotros queremos saber cual es la probabilidad de estar en cada sector después de muchos lanzamientos

```python
import matplotlib.pyplot as plt
import numpy as np
#Agregamos las librerias
```

```python
# Estado inicial
x=1
y=0
z=0
t=0
n=25 #Número de iteraciones
```

```python
# Listas para guardar los valores
x_i=[]
y_i=[]
z_i=[]
t_i=[]
#Hacemos las iteraciones
for i in range(n):
    xi = (1/4)*x + (1/4)*z + (1/2)*t #Calculamos nuevas probabilidades
    yi = (1/2)*x + (1/4)*y + (1/4)*t
    zi = (1/4)*x + (1/2)*y + (1/4)*z
    ti = (1/4)*y + (1/2)*z + (1/4)*t
#Guardamos valores
    x_i.append(xi)
    y_i.append(yi)
    z_i.append(zi)
    t_i.append(ti)

# Actualizamos valores
    x=xi
    y=yi
    z=zi
    t=ti
```

```python
# Graficamos
plt.plot(x_i, label='x_n')
plt.plot(y_i, label='y_n')
plt.plot(z_i, label='z_n')
plt.plot(t_i, label='t_n')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidades')
plt.title('Gráfica')
plt.legend()
plt.grid(True)
plt.show()
```

Conclusión: Despues de muchos lanzamientos nuestra curva se acerca en 0.25 😊

**¿Son todos los secctores igual de probables cuando pase mucho tiempo?**
Si, porque podemos ver que después de muchas iteraciones las probabilidades son las mismas, como se ve en la gráfica
