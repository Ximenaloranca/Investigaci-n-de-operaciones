**Ruta crítica para el desarrollo de una página web**

En este código modelo el proyecto del desarrollo de la pagina web, encontrando su ruta crítica y la red.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definimos las actividades que se tienen que hacer para crear una página web
actividades = {
    "A": {"tiempo en días": 2, "Paso 1": "Definir objetivos"},
    "B": {"tiempo en días": 3, "Paso 2": "Análisis de requerimientos"},
    "C": {"tiempo en días": 4, "Paso 3": "Diseño dr estructura"},
    "D": {"tiempo en días": 5, "Paso 4": "Diseño gráfico"},
    "E": {"tiempo en días": 4, "Paso 5": "Configurar servidor"},
    "F": {"tiempo en días": 6, "Paso 6": "Back end"},
    "G": {"tiempo en días": 6, "Paso 7": "Front end"},
    "H": {"tiempo en días": 5, "Paso 8": "Base de datos"},
    "I": {"tiempo en días": 4, "Paso 9": "Integración"},
    "J": {"tiempo en días": 5, "Paso 10": "Pruebas funcionales"},
    "K": {"tiempo en días": 3, "Paso 11": "Optimización SEO"},
    "L": {"tiempo en días": 3, "Paso 12": "Seguridad/SSL"},
    "M": {"tiempo en días": 4, "Paso 13": "Pruebas de usuario"},
    "N": {"tiempo en días": 2, "Paso 14": "Lanzamiento"},
}
```

```python
# Definimos las dependencias de las actividades que deben de terminar antes de que empiecen otras actividades

dependencias = [
    ("Inicio", "A"),
    ("A", "B"),
    ("B", "C"),
    ("C", "D"), ("C", "F"), ("C", "G"),
    ("B", "H"), ("E", "F"),
    ("F", "I"), ("G", "I"), ("H", "I"),
    ("I", "J"),
    ("J", "K"), ("J", "L"),
    ("K", "M"), ("L", "M"),
    ("M", "N"),
    ("N", "Fin")
]
```

```python

# Creamos el la red dirigida
G = nx.DiGraph()
G.add_edges_from(dependencies)
```

```python

# Agregamos las posiciones de los nodos manualmente para que se vean bonitas

posiciones = {
    "Inicio": (5, 11),
    "A": (5, 10),
    "B": (5, 9),
    "C": (5, 8),
    "D": (3, 7), "F": (5, 7), "G": (7, 7),
    "H": (9, 7), "E": (1, 7),
    "I": (5, 6),
    "J": (5, 5),
    "K": (3, 4), "L": (7, 4),
    "M": (5, 3),
    "N": (5, 2),
    "Fin": (5, 1)
}
```

```python
# Dibujamos la red
plt.figure(figsize=(12,10))
nx.draw(G, pos, with_labels=True, node_size=2800, node_color="lightpink",
        font_size=9, font_weight="bold", arrows=True)

# Etiquetamos los nodos solo con su número
labels = {node: str(node) for node in G.nodes}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=9, font_weight="bold")

plt.title("Red de Proyecto Web", fontsize=14, fontweight="bold")
plt.show()
```

```python

# Calculamos la tuta crítica
durations = {n: activities[n]["duration"] for n in activities}
critical_path = nx.algorithms.dag.dag_longest_path(G, weight="duration")
critical_duration = sum(durations.get(n,0) for n in critical_path)

print("Ruta crítica:", " -> ".join(critical_path))
print("Duración total del proyecto:", critical_duration, "días")
```
